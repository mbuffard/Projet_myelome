<h2> Recherche des drogues ciblés sur les genes dans la base de donnée TTD </h2>
Compte le nombre de drogues pour un gene donné

```python
# Fonction pour rechercher les informations du gène
FILE_TTD="/home/buffard/code/projet_myelome/data/druggable_targets/TTD_DB/P1-01-TTD_target_download_test.txt"
def search_gene(gene_name):
    """
    Searches for gene information in a file based on a given gene name.

    Parameters:
        gene_name (str): The name of the gene to search for.
        exact_match (bool): Whether to search for an exact match of the gene name.

    Returns:
        list or str: A list of target information if a match is found, or a string indicating no results were found.

    Raises:
        FileNotFoundError: If the file specified by FILE does not exist.
    """
    try:
        # Lire le fichier
        with open(FILE_TTD, 'r') as f:
            lines = f.readlines()
        
        results = []
        
        gene_pattern = re.compile(rf"GENENAME\t.*{gene_name}$")

        # Rechercher les lignes contenant le nom de gène exact avec un préfixe ou suffixe quelconque
        for line in lines:
            if gene_pattern.search(line):
                identifier = line.split('\t')[0]
                target_info = [l.strip() for l in lines if l.startswith(identifier) and not ("SEQUENCE" in l or "FUNCTION" in l or "SYNONYMS" in l or "PDBSTRUC" in l or "BIOCLASS" in l or "ECNUMBER" in l)]
                results.append(target_info)
        

        if not results:
            return 0
        
        return results
    except FileNotFoundError:
        return f"Le fichier {FILE_TTD} n'existe pas."
```

```python
count_drug=[]
for node in df['Name']:
    count=0
    output=search_gene(node)
    if not output:
        count_drug.append(0)
    else:
        for idx, target_info in enumerate(output):
        
            drug_info_data = []

            for info in target_info:
                info_tab = info.split('\t')
                if info_tab[1] == "DRUGINFO":
                    count+=1
        count_drug.append(count)

df['drug_count']=count_drug
```

<h2> Recherche des dépendances à un gene dans depmap des lignées de myelome </h2>
On regarde les valeurs dans depmap et on ajoute une colonne avec 3 valeurs "gene absent", T-statistic 'positive' ou 'positive' ou "udefined" si le CRISPR et RNAi sont de signe contraire

```python
FILE_dependencies="../../../../data/depmap/dependencies enriched in multiple_myeloma.csv"

df_dep=pd.read_csv(FILE_dependencies)
df_dep=df_dep[df_dep['Type'] != 'compound']
dep_result=[]
for node in df['Name']:

    if df_dep[df_dep['Gene/Compound']==node].empty:
        dep_result.append("gene absent")
    else:
        sign="unknown"
        for stat in df_dep[df_dep['Gene/Compound']==node]["T-Statistic"]:
            #check if all stat are negative or positive if one stat is positive and other negative sign="unknown"
            if stat<0:
                if sign=="unknown" or sign=="negative":
                    sign="negative"
                else:
                    sign="undefined"
            elif stat>0:
                if sign=="unknown" or sign=="positive":
                    sign="positive"
                else:
                    sign="undefined"
                
        dep_result.append(sign)

df['dep_result']=dep_result
        
```

<h2> Recherche des codépendances entre les genes dans depmap </h2>

Filtre les genes pour lesquels il n'y a aucune drogue

```python
df=pd.read_csv('bad_genes_analysis.csv',sep=',')
```

```python
df=df[df['drug_count']!=0]
df
```

```python
import urllib.request
sep_csv="\t"
sep_genes=";"
gene_list=df['Name'].tolist()
dico_crispr={}
dico_rnai={}
for i in range(0,len(gene_list)):
    dico_crispr[gene_list[i]]=[]
    dico_rnai[gene_list[i]]=[]
    #download file from url
    url="https://depmap.org/portal/gene/"+gene_list[i]+"/top_correlations?dataset_name=Chronos_Combined"
    f=urllib.request.urlopen(url)
    #read the file
    f=f.read().decode('utf-8')

    line=f.split('\n')
    for j in range(1,len(line)):
        gene = line[j].split(",")[0]
        if gene in gene_list:
            dico_crispr[gene_list[i]].append(gene) #add gene
    
    
    #download file from url
    
    url="https://depmap.org/portal/gene/"+gene_list[i]+"/top_correlations?dataset_name=RNAi_merged"
    f=urllib.request.urlopen(url)
    #read the file
    f=f.read().decode('utf-8')

    line=f.split('\n')
    for j in range(1,len(line)):
        gene = line[j].split(",")[0]
        if gene in gene_list:
            dico_rnai[gene_list[i]].append(gene) #add gene
            
    
    
```

```python
#add column in df named "Crispr correlation" with the value associated with key==name in df
for gene in df['Name']:
    df.loc[df['Name']==gene,'Crispr correlation']=sep_genes.join(dico_crispr[gene])
    df.loc[df['Name']==gene,'RNAi correlation']=sep_genes.join(dico_rnai[gene])
df.to_csv('bad_genes_analysis_correlation.csv',index=False,sep=sep_csv)

```
