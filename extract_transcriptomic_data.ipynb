<h1>Extraction des données de transcriptomique</h1>
données issues de "RNA-sequencing data-driven dissection of human plasma cell
differentiation reveals new potential transcription regulators" Leukemia
Les identifiants ensembl ont été mappé avec les identifiants des proteines issues de UniProtKB/Swiss-Prot. 

1) Importation des données de RNAseq

```python
import pandas as pd
df = pd.read_csv('count.normalized_uniprot_mapped.csv',sep="\t")
df
```

```python

df.isnull().sum()
df=df.dropna()
```

```python
df.dtypes
df['uniprot'] = df['uniprot'].astype("string")
df.dtypes
```

2) Filtrage des données


```python
#Elimination des proteines dont le count est <=64 
threshold=64
for i in range(2,len(df.columns)-1):
    cell_line=df.columns[i]
    df[df[cell_line]<=threshold]['uniprot'].to_csv('uniprot_not_detected/'+cell_line+'.csv', index=False)
    print(df[df[cell_line]<=threshold]['uniprot'])
```

On réccupère la liste de tous les noeuds présents dans le réseau de connaissance

```python
df_network = pd.read_csv('all__network_nodes.tsv',sep="\t")
df_network
```

On filtre les séquences codantes ou qui ont un équivalent uniprot et qui sont présentes dans le réseau de connaissance de départ
on genère un fichier pour chaque lignée


```python
list_id=list(df_network['PID'])
df=df[df['uniprot'].isin(list_id)]
threshold=64
f=open('cell_line_with_RNA.csv','w')
for i in range(2,len(df.columns)):
    cell_line=df.columns[i]
    f.write(cell_line+'\t')
    df[df[cell_line]<=threshold]['uniprot'].to_csv('uniprot_not_detected_filtered_uniprot/'+cell_line+'.csv', index=False)
    print(df[df[cell_line]<=threshold]['uniprot'])
f.close()
print(df[cell_line])
```
